## PART A

<u><b>How to classify (anti)neutrino events</b></u>

### Short wide fuzzy shower --> ELECTRON

We usually see shorter, fuzzier tracks. The electron passes through detector and causes ionisation along its path.



### Long straight track --> MUON 
Typically for these event types we see long, straight tracks, which is the muon passing through the detector (~ 1 cell thickness)





### Short tracks/showers with a gap --> NEUTRAL CURRENT

We can't distinguish the neutrino flavor in a Neutral Current event, but we often see short tracks or showers, often with gaps.



### Vertical muon (or activity near edges) --> COSMIC RAY

Cosmic rays come from the Sun and other stars. They can create long tracks that originate outside the detector.

## PART B

<u><b>How to measure the muon (anti)neutrino energy</b></u>

As we cannot see the (anti)neutrino itself, we can only measure its energy by using
the Law of Energy Conservation: we measure the energy of all of the products in the event and
assume that all energy came from the (anti)neutrino.

### 1. <b><i>Measure the muon track length</i>:</b> 

First you measure the length of the muon track with a ruler (choose the view with the longest length). For the near detector, the scale is 11.5:1000, and for the far detector the scale is 3.1:1000. For example, if you measure 5 cm with your ruler, the muon's actual track length at the near detector will be: 
    $$ \mathbf{5 \; cm \; \times \; \frac{1000}{11.5} = \; 434 \; cm.} $$
    If you repeat the same exercise at the far detector, the muon track length will be: 
    $$ \mathbf{5 \; cm \; \times \; \frac{1000}{3.1} = \; 1613 \; cm.} $$

### 2. <b><i>Measure the muon energy</i>:</b> 

A muon is a Minimum Ionizing Particle; as such, it leaves a very small amount of radiation in the detector when it moves and its energy is well understood. In order to measure the muon energy, we look at the track length and apply the following formula:
    $$ \mathbf{E_ \; = \; 200(MeV)+2(\frac{MeV}{cm}) \; \times \; L(cm)} $$
    For example, for a muon track of length 434 cm:
    $$ \mathbf{E_ = 200+2 \times \; 434  =  1068 \; MeV \; =  1.07 \; GeV}$$

### 3. <b><i>Measure the hadronic energy</i>:</b> 

Measuring the hadronic energy is a lot more difficult, so we will use a "hand-wavey" method. Count all non-zero pixels (<sup>*</sup>in both views!<sup>*</sup>) that are not part of the muon track, and evaluate the hadronic energy as:

   $$ \mathbf{E_{had} \; = \; N_{pixels} \; \times \; 40(MeV) } $$
   
   For example, if there are 13 non-zero pixels, E<sub>had</sub> will be:
   
   $$ \mathbf{13 \; \times \; 40(MeV) \; = \; 520 \; MeV \; = \; 0.52 \; GeV} $$

### 4. <b><i>Estimate the neutrino energy</i>:</b> 
Now, in order to find the neutrino energy, simply sum the muon energy and the hadronic energy:
   $$ \mathbf{E_{neutrino} \; = \; E_ \; + \; E_{had}} $$
